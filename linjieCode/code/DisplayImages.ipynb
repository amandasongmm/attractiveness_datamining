```python
import os
imPath = '../neuralNet/deep-visulization/maxPatternWeighted_backprop/'
ext = '.png'
imList = []
for dirpath, dirnames, filenames in os.walk(imPath):
    for filename in [f for f in filenames if f.endswith(ext)]:
        imList.append(os.path.join(dirpath, filename))
imList.sort()
#print imList

plotIm = []
for img in imList:
    imgName = os.path.basename(img)
    if imgName.startswith(('maxMap')):
        plotIm.append(img)
print plotIm[0]

```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import cv2
%matplotlib inline
fig = plt.figure()
fig.set_size_inches(30, 10)
for i in range(1, len(plotIm)+1):
    a=fig.add_subplot(2,5,i)
    imgPath = plotIm[i-1]
    img = cv2.imread(imgPath)
    imgName = os.path.basename(imgPath)
    titleName = imgName[7:-4]
    imgplot = plt.imshow(img)
    a.set_title(titleName,fontsize = 24)
    a.get_xaxis().set_visible(False)
    a.get_yaxis().set_visible(False)

```
