```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import os

#read files
def readFile(fName):
    text_file = open(fName, "r")
    lines = text_file.read().split('\n')
    im = []
    rating = []
    for l in range(len(lines)):
        if len(lines[l]) >2:
            im.append( lines[l].split('.jpg ')[0]+'.jpg')
            rating.append(lines[l].split('.jpg ')[1])
    text_file.close()
    return im, rating
fName = '/home/lli-ms/attractiveness_datamining/linjieCode/data_list/attractive_train.txt'
Im_train, Rating_train = readFile(fName)
print Im_train[1001], Rating_train[1001]
index = np.asarray(range(len(Im_train)))
copy_rating = np.asarray(Rating_train)
maxIndex = index[copy_rating.argsort()[-5:][::-1]]
minIndex = index[copy_rating.argsort()[:5]]
copy_rating.sort()
maxRating = copy_rating[-5:][::-1].astype(float)
minRating = copy_rating[:5].astype(float)
print '**************************Result of max and min on training set****************'
print 'maxIndex: ',maxIndex
print 'minIndex: ',minIndex
print 'maxRating: ',maxRating
print 'minRating: ',minRating
print '****************************************************************************************'


```
