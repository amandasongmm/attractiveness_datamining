```python
import numpy as np
```

```python
all_data = np.load('model_visual_evaluation/test.npz')
```

```python
all_data['label']
```

```python
all_data['outputs']
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(all_data['label'], all_data['outputs'], 'b.')
plt.xlabel('Human rating')
plt.ylabel('Model prediction')
plt.grid(True)
plt.title('')
plt.savefig('test.png')
plt.show()
```

```python
np.corrcoef(all_data['outputs'].reshape((len(all_data['outputs']),)) ,all_data['label'])
```

```python
cor = np.corrcoef(all_data['outputs'].reshape((len(all_data['outputs']),)) ,all_data['label'])[0,1]
```

```python
cor
```

```python
plt.plot(all_data['label'], all_data['outputs'], 'b.')
plt.xlabel('Human rating')
plt.ylabel('Model prediction')
plt.grid(True)
plt.xlim(2, 9)
plt.ylim(2, 9)
plt.gca().set_aspect('equal', adjustable='box')
title_str = 'correlation {0:.2f}'.format(cor)
plt.title(title_str)
plt.show()
```
